```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import split, explode, lower, regexp_extract
```

```python
spark = (SparkSession
         .builder
         .appName("Analyse vocabulariy of Pride and Prejudice.")
         .getOrCreate())
```

```python
book = spark.read.text("../Data/gutenberg_books/1342-0.txt")

lines = book.select(split(book.value, " ").alias('line'))
words = lines.select(explode(lines.line).alias('word'))
words_lower = words.select(lower(words.word).alias('word'))
words_clean = words_lower.select(regexp_extract(words_lower.word, "[a-z]+", 0).alias('word'))
words_non_null = words_clean.filter(words_clean.word != "")
results = words_non_null.groupBy("word").count()
results.orderBy("count", ascending=False).show(10)
```
